```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('ha01_data01.xls')
df.head()
```

```python
plt.rcParams['figure.figsize'] = [10, 5]; 
sns.heatmap(df.corr(), annot = True);
```

```python
df.plot.scatter('tenure', 'earnings', grid=True, title='Scatter plot of Earnings and Tenure');
```

```python
lr1 = sns.lmplot(x = 'tenure', y = 'earnings', data=df, fit_reg = True, ci = None) 
lr1.fig.set_size_inches(10,5);
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('tenure ~ earnings', data=df).fit()

print(model.summary())
```

```python
fig = plt.figure(figsize=(12,8))
fig = sm.graphics.plot_regress_exog(model, 'earnings', fig=fig)
```
